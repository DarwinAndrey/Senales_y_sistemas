```python
# configuraci√≥n del archivo para el Punto 1
%%writefile pages/punto1.py

# Importaciones esenciales para el an√°lisis de sistemas
import streamlit as st
import numpy as np
import scipy.signal as signal
import matplotlib.pyplot as plt
import pandas as pd

# --- Configuraci√≥n general de la P√°gina del Punto 1 ---
st.set_page_config(
    page_title="An√°lisis de Sistemas Din√°micos",  # T√≠tulo descriptivo
    page_icon="üìä",  # √çcono de gr√°fica o an√°lisis
    layout="wide", # Usar el ancho completo para m√°s espacio
    initial_sidebar_state="expanded" # Sidebar visible por defecto
)

# --- Estilo de Matplotlib ---
# Configuraci√≥n para un aspecto limpio y moderno
plt.style.use('fast') # Estilo r√°pido y limpio
plt.rcParams['axes.facecolor'] = '#f8f8f8' # Fondo gris claro para los ejes
plt.rcParams['figure.facecolor'] = '#ffffff' # Fondo blanco para la figura
plt.rcParams['grid.color'] = '#cccccc' # Color de la cuadr√≠cula
plt.rcParams['grid.linestyle'] = '--' # Estilo de la cuadr√≠cula
plt.rcParams['grid.linewidth'] = 0.5 # Grosor de la cuadr√≠cula
plt.rcParams['axes.edgecolor'] = '#333333' # Color del borde de los ejes
plt.rcParams['text.color'] = '#333333' # Color del texto
plt.rcParams['xtick.color'] = '#333333' # Color de las marcas del eje x
plt.rcParams['ytick.color'] = '#333333' # Color de las marcas del eje y
plt.rcParams['axes.labelcolor'] = '#333333' # Color de las etiquetas de los ejes
plt.rcParams['axes.titlecolor'] = '#333333' # Color del t√≠tulo de los ejes


# --- Funciones Auxiliares ---

def calcular_parametros_temporales_escalon(tiempo, respuesta_escalon, valor_final_dc):
    ''' Calcula los par√°metros clave de la respuesta al escal√≥n: Tr, Mp, Tp, Ts.'''
    params_dict = {
        'Tiempo de Levantamiento ($t_r$)': 'N/A',
        'Sobre-impulso M√°ximo ($M_p$)': 'N/A',
        'Tiempo de Pico ($t_p$)': 'N/A',
        'Tiempo de Establecimiento ($t_s$)': 'N/A'
    }

    # Si el valor final es cercano a cero, no se calculan la mayor√≠a de los par√°metros
    if abs(valor_final_dc) < 1e-9:
        return params_dict

    # Criterio del 10% y 90% para Tiempo de Levantamiento
    nivel_10 = 0.10 * valor_final_dc
    nivel_90 = 0.90 * valor_final_dc
    try:
        idx_10 = np.where(respuesta_escalon >= nivel_10)[0][0]
        idx_90 = np.where(respuesta_escalon >= nivel_90)[0][0]
        tr = tiempo[idx_90] - tiempo[idx_10]
        params_dict['Tiempo de Levantamiento ($t_r$)'] = f"{tr:.3f} s"
    except (IndexError, TypeError):
        pass # En caso de que no se alcancen los niveles o haya otros errores

    # Sobre-impulso M√°ximo y Tiempo de Pico
    max_valor = np.max(respuesta_escalon)
    if max_valor > valor_final_dc * (1 + 1e-3): # Considerar un peque√±o margen
        sobre_impulso_percent = ((max_valor - valor_final_dc) / valor_final_dc) * 100
        params_dict['Sobre-impulso M√°ximo ($M_p$)'] = f"{sobre_impulso_percent:.2f} %"
        idx_pico = np.argmax(respuesta_escalon)
        tp = tiempo[idx_pico]
        params_dict['Tiempo de Pico ($t_p$)'] = f"{tp:.3f} s"
    else:
         params_dict['Sobre-impulso M√°ximo ($M_p$)'] = "0.00 %" # No hay sobreimpulso significativo


    # Tiempo de Establecimiento (Criterio del 2%)
    tolerancia_establecimiento = 0.02 * abs(valor_final_dc)
    try:
        # Encontrar todos los puntos que est√°n fuera de la banda de tolerancia
        indices_fuera = np.where(np.abs(respuesta_escalon - valor_final_dc) > tolerancia_establecimiento)[0]
        if indices_fuera.size > 0:
            # El tiempo de establecimiento es el tiempo despu√©s del √∫ltimo punto fuera de la banda
            ultimo_indice_fuera = indices_fuera[-1]
            if ultimo_indice_fuera + 1 < len(tiempo):
                ts = tiempo[ultimo_indice_fuera + 1]
                params_dict['Tiempo de Establecimiento ($t_s$)'] = f"{ts:.3f} s"
            else:
                 # Si el √∫ltimo punto todav√≠a est√° fuera, el sistema no se estableci√≥ completamente en el tiempo de simulaci√≥n
                 params_dict['Tiempo de Establecimiento ($t_s$)'] = "> " + f"{tiempo[-1]:.3f} s"
        else:
            # Si nunca sali√≥ de la banda (empez√≥ dentro o lleg√≥ y se qued√≥)
             params_dict['Tiempo de Establecimiento ($t_s$)'] = f"{tiempo[0]:.3f} s"

    except IndexError:
        pass # En caso de errores en el manejo de √≠ndices

    return params_dict


# --- Estructura de la Interfaz ---

st.title("An√°lisis Interactivo de Sistemas de Segundo Orden")
st.markdown("Explore las caracter√≠sticas de sistemas din√°micos ajustando sus par√°metros.")

# Usar columnas para organizar los controles y la informaci√≥n de componentes
col_controles, col_componentes = st.columns([2, 1]) # Columnas con anchos relativos

with col_controles:
    st.header("‚öôÔ∏è Controles del Sistema")

    # Usar columnas para poner la selecci√≥n de tipo de respuesta y lazo lado a lado
    col_tipo_resp, col_tipo_lazo = st.columns(2)

    with col_tipo_resp:
        # Selecci√≥n de tipo de respuesta (radio buttons para visualizaci√≥n r√°pida)
        tipo_respuesta_sel = st.radio(
            "Seleccione el Tipo de Respuesta Deseado:",
            ('Subamortiguada', 'Amortiguamiento Cr√≠tico', 'Sobreamortiguada', 'Inestable'),
            index=1 # Amortiguamiento Cr√≠tico por defecto (cambio de valor inicial)
        )

    with col_tipo_lazo:
        # Selecci√≥n de tipo de lazo (dropdown para ahorrar espacio)
        tipo_lazo_sel = st.selectbox(
            "Seleccione la Configuraci√≥n del Lazo:",
            ('Lazo Abierto', 'Lazo Cerrado'),
            index=1 # Lazo Cerrado por defecto (cambio de valor inicial)
        )


    # Slider para Factor de Amortiguamiento (Œ∂) - Restricci√≥n visual
    valor_zeta_sel = 0.0
    if tipo_respuesta_sel == 'Subamortiguada':
        valor_zeta_sel = st.slider("Factor de Amortiguamiento (Œ∂):", 0.01, 0.99, 0.5, 0.01, help="0 < Œ∂ < 1") # Cambio de valor inicial
    elif tipo_respuesta_sel == 'Amortiguamiento Cr√≠tico':
        valor_zeta_sel = 1.0
        st.markdown("Factor de Amortiguamiento (Œ∂): **1.0** (Amortiguamiento Cr√≠tico)")
    elif tipo_respuesta_sel == 'Sobreamortiguada':
        valor_zeta_sel = st.slider("Factor de Amortiguamiento (Œ∂):", 1.01, 5.0, 2.0, 0.01, help="Œ∂ > 1") # Cambio de valor inicial
    else: # Inestable
        valor_zeta_sel = st.slider("Factor de Amortiguamiento (Œ∂):", -1.0, -0.01, -0.2, 0.01, help="Œ∂ < 0") # Cambio de valor inicial

    # Slider para Frecuencia Natural (œân)
    valor_wn_sel = st.slider("Frecuencia Natural (œân) [rad/s]:", 1.0, 20.0, 10.0, 0.1, help="Frecuencia natural no amortiguada") # Cambio de valor inicial

with col_componentes:
    st.header("üîß Componentes Equivalentes")
    st.markdown("Estimaci√≥n para sistemas mec√°nicos y el√©ctricos.")

    if tipo_respuesta_sel == 'Inestable':
        st.warning("Los sistemas inestables no corresponden a configuraciones pasivas est√°ndar. No se calculan componentes.")
    else:
        # C√°lculo de componentes mec√°nicos (asumiendo masa unitaria)
        masa_unidad = 1.0 # kg
        rigidez_eq = masa_unidad * valor_wn_sel**2 # k = m * wn^2
        amortiguador_eq = 2 * valor_zeta_sel * valor_wn_sel * masa_unidad # c = 2 * zeta * wn * m

        st.subheader("Sistema Mec√°nico (m = 1 kg)")
        st.metric(label="Rigidez (k)", value=f"{rigidez_eq:.2f} N/m")
        st.metric(label="Amortiguador (c)", value=f"{amortiguador_eq:.2f} Ns/m")

        # C√°lculo de componentes el√©ctricos (asumiendo capacitancia base)
        capacitancia_base_uF = 1000 # ¬µF
        capacitancia_base_F = capacitancia_base_uF * 1e-6 # F

        st.subheader(f"Sistema El√©ctrico (C = {capacitancia_base_uF} ¬µF)")
        # Asegurar que los valores sean positivos para R y L en sistemas estables
        if rigidez_eq > 1e-9 and amortiguador_eq > 1e-9:
             inductancia_eq = (masa_unidad/rigidez_eq) / capacitancia_base_F # L = (m/k) / C
             resistencia_eq = (inductancia_eq * rigidez_eq) / amortiguador_eq # R = (L * k) / c
             st.metric(label="Inductancia (L)", value=f"{inductancia_eq * 1e3:.2f} mH") # Mostrar en mH
             st.metric(label="Resistencia (R)", value=f"{resistencia_eq:.2f} Œ©")
        else:
             st.info("C√°lculo de componentes el√©ctricos no aplicable o no real para estos par√°metros.")


st.markdown("---") # Separador visual

# --- L√≥gica del Sistema ---
# Definici√≥n de las funciones de transferencia
# Numerador es el mismo para Lazo Abierto y Cerrado (si el sistema es est√°ndar)
num_tf = [valor_wn_sel**2]
den_ol_tf = [1, 2 * valor_zeta_sel * valor_wn_sel, valor_wn_sel**2]
tf_open_loop = signal.TransferFunction(num_tf, den_ol_tf)

# Denominador Lazo Cerrado (retroalimentaci√≥n unitaria): Den_cl = Den_ol + Num_ol
den_cl_tf = np.polyadd(den_ol_tf, num_tf)
tf_closed_loop = signal.TransferFunction(num_tf, den_cl_tf)

# Seleccionar la funci√≥n de transferencia actual
tf_activa = tf_open_loop if tipo_lazo_sel == 'Lazo Abierto' else tf_closed_loop

st.subheader("Funci√≥n de Transferencia del Sistema Seleccionado")
# Mostrar la funci√≥n de transferencia en formato LaTeX
if tipo_lazo_sel == 'Lazo Abierto':
    st.latex(f"H(s) = \\frac{{{valor_wn_sel**2:.2f}}}{{s^2 + {2*valor_zeta_sel*valor_wn_sel:.2f}s + {valor_wn_sel**2:.2f}}}")
else:
     # Recalcular los coeficientes del denominador de lazo cerrado para mostrarlos en LaTeX
     coef_den_cl = np.polyadd([1, 2 * valor_zeta_sel * valor_wn_sel, valor_wn_sel**2], [valor_wn_sel**2])
     st.latex(f"H_{{CL}}(s) = \\frac{{{valor_wn_sel**2:.2f}}}{{s^2 + {coef_den_cl[1]:.2f}s + {coef_den_cl[2]:.2f}}}")


# Determinaci√≥n de estabilidad para ajustar el tiempo de simulaci√≥n
polos_sistema = tf_activa.poles
es_estable = np.all(np.real(polos_sistema) < 1e-9) # Considerar polos en el eje imaginario como no estrictamente estables

# C√°lculo del vector de tiempo para las simulaciones
tiempo_sim_final = 10.0 # Tiempo por defecto

if es_estable:
    # Estimar tiempo final basado en el polo real m√°s cercano al eje imaginario
    polos_reales_estables = np.real(polos_sistema[np.real(polos_sistema) < -1e-9])
    if polos_reales_estables.size > 0:
        # Tiempo de establecimiento aproximado en 5 constantes de tiempo del polo dominante
        tiempo_sim_final = max(tiempo_sim_final, 5.0 / abs(np.max(polos_reales_estables)))
    else:
        # Si solo hay polos imaginarios puros, el sistema oscila (marginalmente estable)
        tiempo_sim_final = 20.0 # Mostrar m√°s ciclos
else: # Sistema inestable (polos en semiplano derecho)
     tiempo_sim_final = 5.0 # Mostrar la divergencia inicial

# Limitar el tiempo de simulaci√≥n a un rango razonable
tiempo_sim_final = min(max(tiempo_sim_final, 1.0), 50.0) # Entre 1 y 50 segundos
vector_tiempo_simulacion = np.linspace(0, tiempo_sim_final, 4000) # Usar m√°s puntos para mayor resoluci√≥n gr√°fica


# --- Visualizaci√≥n de An√°lisis y Respuestas ---
st.markdown("---")
st.header("üìà Visualizaci√≥n de An√°lisis y Respuestas")

# Usar columnas para poner las opciones de respuesta (Impulso, Escal√≥n, Rampa)
col_respuestas_tabs = st.columns(1)[0] # Usar una sola columna para las pesta√±as

with col_respuestas_tabs:
    # Definir las pesta√±as
    tab_bode_pz_view, tab_impulso_view, tab_escalon_view, tab_rampa_view = st.tabs([
        "üìä Diagramas de Polos/Ceros y Bode",
        "‚ö° Respuesta al Impulso",
        "ü™ú Respuesta al Escal√≥n",
        "üèûÔ∏è Respuesta a la Rampa"
    ])


    # Contenido de la Pesta√±a: Diagramas de Polos/Ceros y Bode
    with tab_bode_pz_view:
        st.subheader("An√°lisis en el Dominio de la Frecuencia y Plano 's'")

        # Usar columnas para mostrar los dos diagramas lado a lado
        col_pz_graph, col_bode_graph = st.columns(2)

        with col_pz_graph:
            st.markdown("#### Diagrama de Polos y Ceros")
            fig_pz, ax_pz = plt.subplots()
            polos, ceros = tf_activa.poles, tf_activa.zeros

            # Dibujar los ceros (marcadores 'o')
            if ceros.size > 0:
                ax_pz.plot(np.real(ceros), np.imag(ceros), 'o', markersize=10, label='Ceros', color='green', markeredgecolor='black')
            # Dibujar los polos (marcadores 'x')
            ax_pz.plot(np.real(polos), np.imag(polos), 'x', markersize=12, mew=3, label='Polos', color='red')

            # Configuraci√≥n de los ejes y la cuadr√≠cula
            ax_pz.set_title("Plano Complejo 's'")
            ax_pz.set_xlabel("Eje Real (œÉ)"); ax_pz.set_ylabel("Eje Imaginario (jœâ)");
            ax_pz.axhline(0, color='black', lw=0.7); ax_pz.axvline(0, color='black', lw=0.7); # Ejes principales
            ax_pz.grid(True, linestyle='--', alpha=0.6);
            ax_pz.legend(loc='upper right', fancybox=True, framealpha=0.9, shadow=True);
            ax_pz.axis('equal') # Escala igual en ambos ejes para perspectiva correcta
            # Ajustar l√≠mites para incluir todos los polos y ceros c√≥modamente
            if polos.size > 0 or ceros.size > 0:
                max_abs = max(np.max(np.abs(polos)), np.max(np.abs(ceros))) if (polos.size > 0 and ceros.size > 0) else (np.max(np.abs(polos)) if polos.size > 0 else np.max(np.abs(ceros)))
                margin = max(1.5, max_abs * 0.2) # A√±adir un margen
                ax_pz.set_xlim([-max_abs - margin, max_abs + margin])
                ax_pz.set_ylim([-max_abs - margin, max_abs + margin])
            else:
                ax_pz.set_xlim([-5, 5]); ax_pz.set_ylim([-5, 5]); # Limites por defecto si no hay polos/ceros

            st.pyplot(fig_pz) # Mostrar la figura en Streamlit

        with col_bode_graph:
            st.markdown("#### Diagrama de Bode (Respuesta en Frecuencia)")
            # Calcular la respuesta en frecuencia del sistema
            frec_w, mag_db, fase_deg = signal.bode(tf_activa)

            # Crear la figura con dos subplots (magnitud y fase)
            fig_bode, (ax_mag, ax_phase) = plt.subplots(2, 1, figsize=(8, 6), sharex=True)

            # Gr√°fica de Magnitud (en dB)
            ax_mag.semilogx(frec_w, mag_db, color='blue', linewidth=2);
            ax_mag.set_ylabel("Magnitud [dB]");
            ax_mag.set_title("Diagrama de Bode");
            ax_mag.grid(True, which="both", linestyle='--', alpha=0.6)

            # Gr√°fica de Fase (en grados)
            ax_phase.semilogx(frec_w, fase_deg, color='orange', linewidth=2);
            ax_phase.set_xlabel("Frecuencia [rad/s]");
            ax_phase.set_ylabel("Fase [grados]");
            ax_phase.grid(True, which="both", linestyle='--', alpha=0.6)

            st.pyplot(fig_bode) # Mostrar la figura en Streamlit


    # Contenido de la Pesta√±a: Respuesta al Impulso
    with tab_impulso_view:
        st.subheader("Respuesta a una Entrada Impulso Unitario (Œ¥(t))")
        st.markdown("Esta gr√°fica muestra c√≥mo reacciona el sistema a una perturbaci√≥n instant√°nea.")

        # Calcular la respuesta al impulso
        tiempo_impulso, respuesta_impulso = signal.impulse(tf_activa, T=vector_tiempo_simulacion)

        # Crear y mostrar la gr√°fica
        fig_imp, ax_imp = plt.subplots(figsize=(10, 5))
        ax_imp.plot(tiempo_impulso, respuesta_impulso, label="Respuesta al Impulso", color='purple', linewidth=2.5)
        ax_imp.set_title("Respuesta al Impulso Unitario");
        ax_imp.set_xlabel("Tiempo (s)"); ax_imp.set_ylabel("Amplitud");
        ax_imp.legend(); ax_imp.grid(True, linestyle='--', alpha=0.6)
        st.pyplot(fig_imp)

        st.info("El an√°lisis de par√°metros temporales est√°ndar (Tr, Mp, Tp, Ts) se realiza t√≠picamente sobre la respuesta al escal√≥n.")


    # Contenido de la Pesta√±a: Respuesta al Escal√≥n
    with tab_escalon_view:
        st.subheader("Respuesta a una Entrada Escal√≥n Unitario (u(t))")
        st.markdown("Observe c√≥mo el sistema responde a una entrada constante que cambia instant√°neamente.")

        # Calcular la respuesta al escal√≥n
        tiempo_escalon, respuesta_escalon = signal.step(tf_activa, T=vector_tiempo_simulacion)

        # Crear y mostrar la gr√°fica
        fig_step, ax_step = plt.subplots(figsize=(10, 5))
        ax_step.plot(tiempo_escalon, respuesta_escalon, label="Respuesta al Escal√≥n", color='darkgreen', linewidth=2.5)

        # Marcar el valor final si el sistema es estable
        if es_estable and respuesta_escalon.size > 0:
            valor_final_estable = respuesta_escalon[-1]
            ax_step.axhline(valor_final_estable, color='gray', linestyle='--', label=f'Valor Final: {valor_final_estable:.3f}')

        ax_step.set_title("Respuesta al Escal√≥n Unitario");
        ax_step.set_xlabel("Tiempo (s)"); ax_step.set_ylabel("Amplitud");
        ax_step.legend(); ax_step.grid(True, linestyle='--', alpha=0.6)
        st.pyplot(fig_step)

        # Mostrar los par√°metros temporales calculados si es estable
        if es_estable and respuesta_escalon.size > 0:
            st.markdown("#### ‚è±Ô∏è Par√°metros Clave de la Respuesta al Escal√≥n:")
            parametros_obtenidos = calcular_parametros_temporales_escalon(tiempo_escalon, respuesta_escalon, valor_final_estable)

            # Mostrar los par√°metros en un formato m√°s legible, quiz√°s usando columnas
            col_p1, col_p2 = st.columns(2)
            with col_p1:
                st.metric(label=list(parametros_obtenidos.keys())[0], value=list(parametros_obtenidos.values())[0])
                st.metric(label=list(parametros_obtenidos.keys())[1], value=list(parametros_obtenidos.values())[1])
            with col_p2:
                st.metric(label=list(parametros_obtenidos.keys())[2], value=list(parametros_obtenidos.values())[2])
                st.metric(label=list(parametros_obtenidos.keys())[3], value=list(parametros_obtenidos.values())[3])

        elif not es_estable:
            st.warning("El sistema es inestable (o marginalmente estable). Los par√°metros temporales est√°ndar de la respuesta al escal√≥n no son aplicables o significativos.")


    # Contenido de la Pesta√±a: Respuesta a la Rampa
    with tab_rampa_view:
        st.subheader("Respuesta a una Entrada Rampa (t ‚ãÖ u(t))")
        st.markdown("Visualice la respuesta del sistema a una entrada que aumenta linealmente con el tiempo.")

        # Calcular la respuesta a la rampa
        # La entrada U para lsim debe ser el vector de tiempo para una rampa unitaria
        tiempo_rampa, respuesta_rampa, _ = signal.lsim(tf_activa, U=vector_tiempo_simulacion, T=vector_tiempo_simulacion)

        # Crear y mostrar la gr√°fica
        fig_rampa, ax_rampa = plt.subplots(figsize=(10, 5))
        ax_rampa.plot(tiempo_rampa, respuesta_rampa, label="Respuesta a la Rampa", color='chocolate', linewidth=2.5)
        ax_rampa.plot(vector_tiempo_simulacion, vector_tiempo_simulacion, '--', label="Entrada Rampa Ideal", color='gray') # Graficar la entrada ideal
        ax_rampa.set_title("Respuesta a la Rampa");
        ax_rampa.set_xlabel("Tiempo (s)"); ax_rampa.set_ylabel("Amplitud");
        ax_rampa.legend(); ax_rampa.grid(True, linestyle='--', alpha=0.6)
        st.pyplot(fig_rampa)

        st.info("La respuesta a la rampa es √∫til para analizar el error de estado estacionario para este tipo de entrada.")
```

```python
# Crear el archivo para el Punto 2 con una nueva implementaci√≥n
%%writefile pages/punto2.py
import streamlit as st
import numpy as np
import scipy.signal as signal
from scipy.fft import fft, fftfreq
import matplotlib.pyplot as plt
from scipy.io.wavfile import read, write
from pydub import AudioSegment
import io
import os
import subprocess
import yt_dlp
import browser_cookie3
import traceback # Importar traceback para depuraci√≥n

# --- Configuraci√≥n de la P√°gina ---
st.set_page_config(
    page_title="Simulador SSB-AM",
    page_icon="üì°",
    layout="wide",
)

# --- Estilo de Matplotlib ---
plt.style.use('seaborn-v0_8-notebook') # Usar un estilo diferente
plt.rcParams['axes.facecolor'] = '#ffffff'
plt.rcParams['figure.facecolor'] = '#ffffff'
plt.rcParams['lines.linewidth'] = 2
plt.rcParams['axes.labelsize'] = 10
plt.rcParams['axes.titlesize'] = 12


# --- Funciones de Procesamiento y Visualizaci√≥n ---

def load_audio_segment(file_path, start_sec, duration_sec, target_fs):
    """Loads a segment of an audio file, converts to mono and target sample rate."""
    try:
        audio = AudioSegment.from_file(file_path)
        audio = audio.set_channels(1)
        audio = audio.set_frame_rate(target_fs)
        start_ms = start_sec * 1000
        end_ms = start_ms + duration_sec * 1000
        segment = audio[start_ms:end_ms]
        samples = np.array(segment.get_array_of_samples()).astype(np.float64)
        # Normalize to avoid clipping in plots/audio players
        if np.max(np.abs(samples)) > 1e-9:
            samples /= np.max(np.abs(samples))
        return target_fs, samples
    except Exception as e:
        st.error(f"Error loading audio segment: {e}")
        return None, None

@st.cache_data
def download_youtube_audio(url, target_fs=48000):
    """Downloads audio from YouTube and processes it."""
    temp_filename = "temp_downloaded_audio"
    output_filename = "youtube_audio.wav"

    try:
        # Use yt-dlp to download audio
        ydl_opts = {
            'format': 'bestaudio/best',
            'outtmpl': f'{temp_filename}.%(ext)s',
            'quiet': True,
            'no_warnings': True,
        }
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            info_dict = ydl.extract_info(url, download=True)
            downloaded_file_ext = info_dict.get('ext', 'm4a') # Default to m4a if ext not found
            downloaded_file_path = f'{temp_filename}.{downloaded_file_ext}'

        # Convert downloaded audio to WAV using pydub
        audio = AudioSegment.from_file(downloaded_file_path)
        audio = audio.set_channels(1) # Ensure mono
        audio = audio.set_frame_rate(target_fs) # Resample to target fs

        audio.export(output_filename, format="wav")

        # Clean up temporary file
        if os.path.exists(downloaded_file_path):
             os.remove(downloaded_file_path)

        return output_filename
    except Exception as e:
        st.error(f"Error downloading or processing YouTube audio: {e}")
        return None


def ssb_modulation(message_signal, fc, fs, sideband='USB'):
    """Performs SSB-AM modulation."""
    t = np.linspace(0, len(message_signal)/fs, len(message_signal), endpoint=False)
    hilbert_transform = signal.hilbert(message_signal)
    m_hat_t = np.imag(hilbert_transform) # Quadrature component

    carrier = np.cos(2 * np.pi * fc * t)
    quadrature_carrier = np.sin(2 * np.pi * fc * t)

    if sideband == 'USB':
        # s_USB(t) = m(t)cos(2*pi*fc*t) - m_hat(t)sin(2*pi*fc*t)
        ssb_signal = message_signal * carrier - m_hat_t * quadrature_carrier
    else: # LSB
        # s_LSB(t) = m(t)cos(2*pi*fc*t) + m_hat(t)sin(2*pi*fc*t)
        ssb_signal = message_signal * carrier + m_hat_t * quadrature_carrier

    return ssb_signal

def ssb_demodulation(ssb_signal, fc, fs, filter_cutoff_freq):
    """Performs SSB-AM coherent demodulation and LPF."""
    t = np.linspace(0, len(ssb_signal)/fs, len(ssb_signal), endpoint=False)
    # Coherent demodulation (multiplication by local carrier)
    local_carrier = np.cos(2 * np.pi * fc * t)
    demodulated_signal = ssb_signal * local_carrier

    # Low-pass filter to extract the baseband signal
    nyquist = 0.5 * fs
    # Ensure cutoff frequency is valid
    if filter_cutoff_freq >= nyquist or filter_cutoff_freq <= 0:
        st.error(f"Frecuencia de corte del filtro ({filter_cutoff_freq} Hz) no es v√°lida. Debe ser > 0 y < {nyquist} Hz.")
        return None, (None, None)

    try:
        # Design Butterworth filter
        b_lpf, a_lpf = signal.butter(6, filter_cutoff_freq / nyquist, btype='low')
        # Apply the filter
        recovered_signal = signal.lfilter(b_lpf, a_lpf, demodulated_signal)
        return recovered_signal, (b_lpf, a_lpf)
    except Exception as e:
        st.error(f"Error durante el dise√±o o aplicaci√≥n del filtro pasa-bajas: {e}")
        st.code(traceback.format_exc()) # Mostrar traceback para depuraci√≥n
        return None, (None, None)


def plot_time_and_freq(t, signal_data, fs, title, freq_xlim=None):
    """Plots signal in time and frequency domains side-by-side."""
    col_time, col_freq = st.columns(2)
    with col_time:
        st.markdown(f"**{title} (Dominio del Tiempo)**")
        fig_t, ax_t = plt.subplots(figsize=(6, 3))
        ax_t.plot(t, signal_data)
        ax_t.set_xlabel("Tiempo (s)")
        ax_t.set_ylabel("Amplitud")
        ax_t.grid(True)
        st.pyplot(fig_t)

    with col_freq:
        st.markdown(f"**{title} (Dominio de la Frecuencia)**")
        n = len(signal_data)
        yf = fft(signal_data)
        xf = fftfreq(n, 1/fs)
        fig_f, ax_f = plt.subplots(figsize=(6, 3))
        ax_f.plot(np.fft.fftshift(xf), 2.0/n * np.abs(np.fft.fftshift(yf)))
        ax_f.set_xlabel("Frecuencia (Hz)")
        ax_f.set_ylabel("Magnitud")
        if freq_xlim is not None:
            ax_f.set_xlim([-freq_xlim, freq_xlim])
        ax_f.grid(True)
        st.pyplot(fig_f)

def plot_filter_characteristics(b, a, fs, title):
    """Plots filter Bode plot and pole-zero diagram."""
    if b is None or a is None:
        st.warning(f"No se pudo graficar las caracter√≠sticas del filtro '{title}' debido a un error anterior.")
        return

    col_bode, col_pz = st.columns(2)

    with col_bode:
        st.markdown(f"##### Diagrama de Bode ({title})")
        try:
            w, h = signal.freqz(b, a, worN=8000)
            frecuencias_hz = 0.5 * fs * w / np.pi
            fig_bode, ax_bode = plt.subplots(figsize=(6, 4))
            ax_bode.plot(frecuencias_hz, 20 * np.log10(abs(h) + np.finfo(float).eps), 'b')
            ax_bode.set_ylabel('Amplitud [dB]', color='b')
            ax_bode.set_xlabel('Frecuencia [Hz]')
            ax_bode.set_title("Magnitud")
            ax_bode.grid(True)
            st.pyplot(fig_bode)
        except Exception as e:
            st.error(f"Error al graficar el diagrama de Bode para '{title}': {e}")
            st.code(traceback.format_exc())


    with col_pz:
        st.markdown(f"##### Plano de Polos y Ceros ({title})")
        try:
            z, p, k = signal.tf2zpk(b, a)
            fig_pz, ax_pz = plt.subplots(figsize=(4, 4))
            circulo_unitario = plt.Circle((0,0), 1, fill=False, color='gray', ls='--')
            ax_pz.add_artist(circulo_unitario)
            if z.size > 0:
                ax_pz.plot(np.real(z), np.imag(z), 'o', markersize=8, fillstyle='none', label='Ceros')
            if p.size > 0:
                ax_pz.plot(np.real(p), np.imag(p), 'x', markersize=8, label='Polos')

            ax_pz.set_aspect('equal')
            ax_pz.set_xlim([-1.2, 1.2])
            ax_pz.set_ylim([-1.2, 1.2])
            ax_pz.set_xlabel("Eje Real")
            ax_pz.set_ylabel("Eje Imaginario")
            ax_pz.grid(True)
            ax_pz.legend(fancybox=True, framealpha=1, shadow=True, borderpad=1)
            st.pyplot(fig_pz)
        except Exception as e:
             st.error(f"Error al graficar el plano de Polos/Ceros para '{title}': {e}")
             st.code(traceback.format_exc())



# --- Interfaz de Usuario Principal ---

st.title("Experimentaci√≥n con Modulaci√≥n y Demodulaci√≥n SSB-AM")
st.markdown("Seleccione una se√±al mensaje y configure los par√°metros de transmisi√≥n para visualizar el proceso.")

# Controles en la barra lateral
# Mover controles de la barra lateral al √°rea principal, usando columnas
st.header("üéõÔ∏è Configuraci√≥n de la Simulaci√≥n")

col_source, col_params = st.columns(2)

with col_source:
    message_source = st.selectbox(
        "Origen de la Se√±al Mensaje:",
        ('Pulso Rectangular', 'Audio de YouTube', 'Subir Archivo de Audio'),
        index=1 # Cambiar valor inicial
    )

with col_params:
     carrier_freq = st.slider("Frecuencia de la Portadora (fc) [Hz]", 4000, 15000, 8000, 100) # Cambiar valor inicial
     selected_sideband = st.radio(
         "Banda Lateral a Transmitir:",
         ('Banda Lateral Superior (USB)', 'Banda Lateral Inferior (LSB)'),
         index=1 # Cambiar valor inicial
     )
     sample_rate = 48000 # Hz
     analysis_duration = 5 # seconds
     filter_cutoff = st.slider("Frecuencia de Corte del LPF [Hz]", 1000, min(carrier_freq - 500, sample_rate // 2 - 1000), 3000, 50) # Cambiar valor inicial
     if filter_cutoff >= carrier_freq:
         st.warning("La frecuencia de corte del filtro pasabajas debe ser menor que la frecuencia de la portadora.")


message_signal = None
time_vector = None

# Cargar la se√±al mensaje seg√∫n la selecci√≥n del usuario (esta l√≥gica permanece en el √°rea principal)
if message_source == 'Pulso Rectangular':
    st.info("Generando pulso rectangular de 1 segundo...")
    time_vector = np.linspace(0, analysis_duration, analysis_duration * sample_rate, endpoint=False)
    message_signal = np.zeros_like(time_vector)
    pulse_start = 0.5 # Cambiar valor inicial
    pulse_end = 1.5 # Cambiar valor inicial
    message_signal[int(pulse_start * sample_rate):int(pulse_end * sample_rate)] = 1.0

elif message_source == 'Audio de YouTube':
    youtube_url = st.text_input("Enlace de YouTube:", "https://www.youtube.com/watch?v=q_q61B-DyPk") # Cambiar valor inicial
    if youtube_url:
        audio_file_path = download_youtube_audio(youtube_url, sample_rate)
        if audio_file_path:
            try:
                full_fs, full_audio_data = read(audio_file_path)
                if full_audio_data.ndim > 1:
                    full_audio_data = full_audio_data.mean(axis=1)

                total_duration_sec = len(full_audio_data) / full_fs
                max_start_sec = int(total_duration_sec - analysis_duration)

                if max_start_sec < 0:
                    st.warning(f"El audio es m√°s corto que {analysis_duration} segundos. Usando el audio completo.")
                    segment_data = full_audio_data
                else:
                    start_analysis_sec = st.number_input(
                        f"Segundo de inicio para el segmento de {analysis_duration}s:",
                        min_value=0,
                        max_value=max_start_sec,
                        value=min(60, max_start_sec), # Cambiar valor inicial
                        step=1
                    )
                    start_sample = int(start_analysis_sec * full_fs)
                    end_sample = int(start_sample + analysis_duration * full_fs)
                    segment_data = full_audio_data[start_sample:end_sample]

                message_signal = segment_data.astype(np.float64)
                 # Normalize
                if np.max(np.abs(message_signal)) > 1e-9:
                     message_signal /= np.max(np.abs(message_signal))

                time_vector = np.linspace(0, len(message_signal)/full_fs, len(message_signal), endpoint=False)

                st.audio(message_signal, sample_rate=full_fs)
            except Exception as e:
                st.error(f"Error processing downloaded audio: {e}")

elif message_source == 'Subir Archivo de Audio':
    uploaded_file = st.file_uploader("Suba un archivo .wav o .mp3", type=["wav", "mp3"])
    if uploaded_file:
        try:
            # Save uploaded file temporarily to process with pydub
            with open("uploaded_temp_audio", "wb") as f:
                 f.write(uploaded_file.getbuffer())

            fs_read, audio_data_raw = load_audio_segment("uploaded_temp_audio", 0, 1000, sample_rate) # Load a large chunk initially
            if fs_read is not None:
                 total_duration_sec = len(audio_data_raw) / fs_read
                 max_start_sec = int(total_duration_sec - analysis_duration)

                 if max_start_sec < 0:
                     st.warning(f"El audio es m√°s corto que {analysis_duration} segundos. Usando el audio completo.")
                     segment_data = audio_data_raw
                 else:
                     start_analysis_sec = st.number_input(
                         f"Segundo de inicio para el segmento de {analysis_duration}s:",
                         min_value=0,
                         max_value=max_start_sec,
                         value=0,
                         step=1
                     )
                     # Reload only the required segment efficiently
                     fs_seg, segment_data = load_audio_segment("uploaded_temp_audio", start_analysis_sec, analysis_duration, sample_rate)


                 if segment_data is not None:
                    message_signal = segment_data
                    time_vector = np.linspace(0, len(message_signal)/sample_rate, len(message_signal), endpoint=False)
                    st.audio(message_signal, sample_rate=sample_rate)

            # Clean up temporary file
            if os.path.exists("uploaded_temp_audio"):
                 os.remove("uploaded_temp_audio")

        except Exception as e:
            st.error(f"Error processing uploaded file: {e}")


# --- Visualizaci√≥n del Proceso ---

if message_signal is not None and time_vector is not None:
    st.markdown("---")

    st.header("Etapas del Proceso SSB-AM")

    # Usar pesta√±as para organizar las etapas
    tab_mensaje, tab_modulacion, tab_demodulacion_mult, tab_filtro, tab_recuperacion = st.tabs([
        "Etapa 1: Mensaje Original",
        "Etapa 2: Se√±al Modulada",
        "Etapa 3: Demodulaci√≥n (Multiplicaci√≥n)",
        "Etapa 4: Filtro Pasa-Bajas",
        "Etapa 5: Se√±al Recuperada"
    ])


    with tab_mensaje:
        st.subheader("Se√±al Mensaje Original")
        plot_time_and_freq(time_vector, message_signal, sample_rate, "Se√±al Mensaje", freq_xlim=4000)

    with tab_modulacion:
        st.subheader(f"Se√±al Modulada ({selected_sideband})")
        ssb_output = ssb_modulation(message_signal, carrier_freq, sample_rate, selected_sideband.split(" ")[-1][:-1]) # Extract 'USB' or 'LSB'
        plot_time_and_freq(time_vector, ssb_output, sample_rate, "Se√±al Modulada", freq_xlim=carrier_freq + 4000)

    with tab_demodulacion_mult:
        st.subheader("Se√±al Demodulada (Post-Multiplicaci√≥n)")
        temp_demod_mult = ssb_output * np.cos(2 * np.pi * carrier_freq * time_vector)
        plot_time_and_freq(time_vector, temp_demod_mult, sample_rate, "Se√±al Post-Multiplicaci√≥n", freq_xlim=2*carrier_freq + 4000)

    with tab_filtro:
        st.subheader("Filtro Pasa-Bajas de Recuperaci√≥n")
        # Design Butterworth filter
        recovered_signal, (lpf_b, lpf_a) = ssb_demodulation(ssb_output, carrier_freq, sample_rate, filter_cutoff)
        plot_filter_characteristics(lpf_b, lpf_a, sample_rate, "Caracter√≠sticas del Filtro LPF")

    with tab_recuperacion:
        st.subheader("Se√±al Mensaje Recuperada")
        col_recovered_plot, col_recovered_audio = st.columns(2)
        with col_recovered_plot:
            st.markdown("**Comparaci√≥n: Original vs. Recuperada**")
            fig_comp, ax_comp = plt.subplots(figsize=(6, 3))
            ax_comp.plot(time_vector, message_signal, label="Original", alpha=0.7)
            ax_comp.plot(time_vector, recovered_signal, label="Recuperada", linestyle='--')
            ax_comp.set_xlabel("Tiempo (s)"); ax_comp.set_ylabel("Amplitud");
            ax_comp.legend(); ax_comp.grid(True)
            st.pyplot(fig_comp)

        with col_recovered_audio:
            st.markdown("**Audio Recuperado:**")
            try:
                wav_buffer = io.BytesIO()
                if np.max(np.abs(recovered_signal)) > 1e-9:
                     audio_int16 = (recovered_signal / np.max(np.abs(recovered_signal)) * 32767).astype(np.int16)
                else:
                     audio_int16 = (recovered_signal * 32767).astype(np.int16)

                write(wav_buffer, sample_rate, audio_int16)
                st.audio(wav_buffer, format='audio/wav')
            except Exception as e:
                st.error(f"Error generating audio: {e}")


else:
    st.info("Seleccione una fuente de se√±al mensaje en la configuraci√≥n para comenzar.")
```
